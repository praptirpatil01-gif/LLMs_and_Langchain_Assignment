## Question 1: What are Large Language Models (LLMs) and how do they function?

Large Language Models (LLMs) are advanced Artificial Intelligence (AI) models trained on massive amounts of text data. They use deep learning, particularly the Transformer architecture, to understand, generate, summarize, and translate human language.

LLMs learn patterns, grammar, context, and relationships between words during training. When a user provides a prompt, the model predicts the most probable next word repeatedly until it generates a complete response.

### How LLMs Function
* Collect large text datasets.
* Tokenize the text into smaller units.
* Train using deep neural networks.
* Learn contextual relationships between words.
* Generate responses by predicting the next token.

### Examples of LLMs
* GPT-4/GPT-5
* LLaMA
* Gemini
* Claude
* Mistral

### Applications
* Chatbots
* Text summarization
* Code generation
* Translation
* Question Answering
* Content creation

## Question 5: Compare and contrast LangChain and LlamaIndex. What unique problems does each solve?

| LangChain | LlamaIndex |
| :-------------------------------- | :------------------------------------------- |
| * Framework for building LLM applications | * Framework for indexing and retrieving documents |
| * Connects LLMs with external tools       | * Connects LLMs with private data            |
| * Supports memory and agents             | * Supports document indexing                 |
| * Good for chatbots                     | * Good for document Q&A                      |
| * Integrates APIs                       | * Integrates local documents                 |
| * Supports workflows                    | * Supports Retrieval-Augmented Generation (RAG) |

### LangChain Solves
* Prompt chaining
* Agent creation
* Tool integration
* Workflow automation
* Memory management

### LlamaIndex Solves
* Document indexing
* Semantic search
* Knowledge retrieval
* Private document Q&A
* Retrieval-Augmented Generation (RAG)

### Conclusion

LangChain focuses on building intelligent AI workflows, while LlamaIndex specializes in organizing and retrieving information from large document collections.

## Question 6: Implement a basic LangChain pipeline using OpenAI's LLM

```python
!pip install -q transformers torch

from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="distilgpt2"
)

prompt = "Question: What is Artificial Intelligence?\nAnswer:"

result = generator(
    prompt,
    max_length=80,
    do_sample=True
)

print(result[0]["generated_text"])
```

## Question 8: Create a LlamaIndex implementation

```python
!pip -q install langchain langchain-community transformers sentence-transformers faiss-cpu pypdf huggingface_hub accelerate
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
docs=['Python is a programming language.','LangChain connects LLM components.']
emb=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
db=FAISS.from_texts(docs,emb)
print(db.similarity_search('What is LangChain?',k=1)[0].page_content)
```
